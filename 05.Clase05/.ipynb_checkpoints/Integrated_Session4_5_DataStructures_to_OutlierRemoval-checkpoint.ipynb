## 1. Python lists: simple data storage

Lists are useful for storing ordered values and doing quick exploratory calculations.
In this mining example, you are given daily gold prices.


```python
# Precios diarios del oro en USD/oz
gold_prices = [1930.5, 1942.3, 1928.4, 1950.0, 1935.6, 1962.1, 1948.7]

# TODO 1: mostramos el numero de observaciones
print("Numero de observaciones: ", len(gold_prices))

# TODO 2: calculamos la media o promedio
average_price = sum(gold_prices) / len(gold_prices)
print("Media o promedio:", average_price)

# TODO 3: agregamos nuevo valor: 1955.3
gold_prices.append(1955.3)
print("Lista con nuevo precio:", gold_prices)

# TODO 4: ordenamos lista
gold_prices.sort()
print("Lista ordenada:", gold_prices)

# TODO 5: imprimimos el valor minimo y maximo
print("Mayor: ", max(gold_prices))
print("Menor: ", min(gold_prices))

gold_prices
```

## 2. NumPy arrays: numerical computation

Arrays are more efficient than lists for numerical workflows.  
Here you have ore grade data from three pits.


```python
import numpy as np

# Leyes de mineral de 3 tajos
grades = np.array([
    [0.80, 1.00, 0.90, 1.10, 0.85],   # Tajo A
    [1.20, 1.00, 1.30, 1.10, 1.15],   # Tajo B
    [0.70, 0.65, 0.80, 0.75, 0.72]    # Tajo C
])

# TODO 1: calculamos el promedio de ley por tajo
print("Promedio por tajo:", np.mean(grades, axis=1))

# TODO 2: calculamos el promedio general de todas las leyes
print("Promedio general:", np.mean(grades))

# TODO 3: mostramos la ley minima y maxima
print("Ley minima:", np.min(grades))
print("Ley maxima:", np.max(grades))

# TODO 4: aplicamos una corrección de laboratorio de +0.05 a todos los valores
corrected_grades = grades + 0.05
print("Leyes corregidas:\n", corrected_grades)

grades
```

## 3. pandas DataFrame: tabular data

DataFrames are the workhorse of data science.  
This example contains daily production information for several mining operations.


```python
import pandas as pd

data = {
    "Mine": ["Cerro Azul", "El Dorado", "Zinc Valley", "Copper Hill", "Lithium Lakes"],
    "Mineral": ["Gold", "Gold", "Zinc", "Copper", "Lithium"],
    "Daily_Production_t": [1200, 1350, 800, 950, 700],
    "Ore_Grade_pct": [1.2, 1.1, 4.5, 2.3, 0.7]
}

df = pd.DataFrame(data)

# TODO 1: inspeccionamos el dataset
print(df.head())
print(df.describe())

# TODO 2: creamos una nueva columna llamada Metal_Content_kg
# Fórmula: produccion_t * ley_pct * 10
df["Metal_Content_kg"] = df["Daily_Production_t"] * df["Ore_Grade_pct"] * 10
print(df)

# TODO 3: filtramos filas con Ore_Grade_pct > 1.0
high_grade_df = df[df["Ore_Grade_pct"] > 1.0]
print(high_grade_df)

# TODO 4: calculamos el promedio de produccion diaria por mineral
grouped = df.groupby("Mineral")["Daily_Production_t"].mean()
print(grouped)

df
```

## 4. Basic data cleaning: duplicates, text uniformity, and missing production

Before advanced preprocessing, it is useful to clean obvious issues:
- duplicated rows
- inconsistent text labels
- missing values in a single column


```python
import pandas as pd

messy_data = {
    "Mine_ID": ["mn001", "MN001", "MN002", "MN002", " MN003 "],
    "Mineral": ["gold", "Gold ", "Copper", "copper", " Silver"],
    "Production": [1000, 1000, 850, 850, None]
}

df_messy = pd.DataFrame(messy_data)

# TODO 1: inspeccionamos el dataset original
print(df_messy)

# TODO 2: estandarizamos el texto (mayusculas y espacios)
df_clean = df_messy.copy()
df_clean["Mine_ID"] = df_clean["Mine_ID"].str.strip().str.upper()
df_clean["Mineral"] = df_clean["Mineral"].str.strip().str.capitalize()

# TODO 3: eliminamos duplicados luego de limpiar el texto
df_clean = df_clean.drop_duplicates()

# TODO 4: rellenamos valores faltantes de Production con 0
df_clean["Production"] = df_clean["Production"].fillna(df_clean["Production"].median())

# TODO 5: inspeccionamos el resultado limpio
print(df_clean)

df_messy
```

## 6. Outlier detection and removal with IQR

This is the last topic for this session.  
The dataset below contains drillhole gold assays. One value looks suspiciously high.

### Tasks
1. compute Q1, Q3 and IQR for `Au_gpt`  
2. calculate lower and upper bounds  
3. identify the outlier rows  
4. remove outliers  
5. compare the mean before and after filtering


```python
import pandas as pd

assay_data = {
    "Hole_ID": ["DH001", "DH002", "DH003", "DH004", "DH005", "DH006", "DH007", "DH008"],
    "Au_gpt": [1.2, 1.5, 1.4, 1.6, 1.3, 1.5, 1.4, 12.8],
    "Cu_pct": [0.22, 0.25, 0.24, 0.27, 0.23, 0.26, 0.24, 0.29]
}

df_assay = pd.DataFrame(assay_data)

# TODO 1: calculamos los cuartiles y el RIC (Rango Intercuartílico) para Au_gpt
Q1 = df_assay["Au_gpt"].quantile(0.25)
Q3 = df_assay["Au_gpt"].quantile(0.75)
IQR = Q3 - Q1
print(f"Q1: {Q1} | Q3: {Q3} | IQR: {IQR}")

# TODO 2: calculamos los limites inferior y superior
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Limite inferior: {lower_bound} | Limite superior: {upper_bound}")

# TODO 3: identificamos los outliers
outliers = df_assay[(df_assay["Au_gpt"] < lower_bound) | (df_assay["Au_gpt"] > upper_bound)]
print("Outliers:")
print(outliers)

# TODO 4: eliminamos los outliers
df_assay_clean = df_assay[(df_assay["Au_gpt"] >= lower_bound) & (df_assay["Au_gpt"] <= upper_bound)]

# TODO 5: comparamos las medias antes y despues del filtrado
print("Media antes del filtrado:", df_assay["Au_gpt"].mean())
print("Media despues del filtrado:", df_assay_clean["Au_gpt"].mean())

df_assay
```
