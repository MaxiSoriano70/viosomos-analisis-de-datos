This session introduces Python data structures commonly used in data science workflows: `lists`,`dictionaries` and `NumPy arrays`. We will use examples in the context of mining industry to demonstrate their applications.

## 1. Python Lists
Lists are one of the most versatile data structures in Python. You can store multiple items in a single variable.

```python
# Creating a list of mine names
mines = ['Cascabel', 'Chuquipata', 'Mirador', 'Culebrillas']
print("List of mines:", mines)

# Adding a new mine
mines.append('Cerro Verde')
print("Updated list of mines:", mines)

# Accessing and modifying items
print("First mine in list:", mines[0])
mines[2] = 'Fruta del Norte'
print("Modified list:", mines)
```

## 2. Dictionaries
Dictionaries consist of a collection of key-value pairs. A dictionary is defined by enclosing a comma-separated list of key-value pairs in curly braces, with a colon separating each key from the associated value.

```python
# Create a dictionary with mining production data
data = {
    'Mine': ['Culebrillas', 'Mirador', 'Fruta del Norte', 'Cascabel'],
    'Mineral': ['Copper', 'Copper', 'Gold', 'Gold'],
    'Daily_Production_tonnes': [3200, 4100, 2900, 3700],
    'Ore_Grade_%': [1.1, 0.9, 1.5, 0.8]
}

print(data['Daily_Production_tonnes'])
print(data.keys())
```

## 3. NumPy Arrays
NumPy is used for efficient numerical operations, especially on large datasets.
Like lists, arrays are collections of items, but of the same type (e.g., all numbers or all strings).

```python
import numpy as np

# Ore grades in percentage from different samples
ore_grades = [1.2, 0.9, 1.5, 1.1, 0.8]
grades_array = np.array(ore_grades)
print("Ore Grades Array:", grades_array)

# Perform numerical operations
print("Average ore grade:", grades_array.mean())
print("Standard deviation:", grades_array.std())
```
